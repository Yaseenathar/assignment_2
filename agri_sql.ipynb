```python
##Top 5 Districts by Wheat Yield Increase Over the Last 5 Years
import pandas as pd

# Define the SQL query
query = '''
WITH Yield_Recent AS (
    SELECT
        Dist_Name,
        AVG(WHEAT_YIELD_Kg_per_ha) AS Avg_Recent_Yield
    FROM agri_data
    WHERE Year = (SELECT MAX(Year) FROM agri_data)
    GROUP BY Dist_Name
),
Yield_Past AS (
    SELECT
        Dist_Name,
        AVG(WHEAT_YIELD_Kg_per_ha) AS Avg_Past_Yield
    FROM agri_data
    WHERE Year = (SELECT MAX(Year) FROM agri_data) - 5
    GROUP BY Dist_Name
),
Yield_Change AS (
    SELECT
        r.Dist_Name,
        r.Avg_Recent_Yield,
        p.Avg_Past_Yield,
        (r.Avg_Recent_Yield - p.Avg_Past_Yield) AS Yield_Increase
    FROM Yield_Recent r
    JOIN Yield_Past p ON r.Dist_Name = p.Dist_Name
)
SELECT *
FROM Yield_Change
ORDER BY Yield_Increase DESC
LIMIT 5;
'''

# Execute and fetch into a DataFrame
cursor.execute(query)
results = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
df_top5 = pd.DataFrame(results, columns=columns)

# Display result
print(df_top5)
```

```python
##States with the Highest Growth in Oilseed Production (5-Year Growth Rate)
import pandas as pd

# Define SQL query
query = '''
WITH Recent AS (
    SELECT 
        State_Name, 
        SUM(OILSEEDS_PRODUCTION_1000_tons) AS Recent_Production
    FROM agri_data
    WHERE Year = (SELECT MAX(Year) FROM agri_data)
    GROUP BY State_Name
),
Past AS (
    SELECT 
        State_Name, 
        SUM(OILSEEDS_PRODUCTION_1000_tons) AS Past_Production
    FROM agri_data
    WHERE Year = (SELECT MAX(Year) FROM agri_data) - 5
    GROUP BY State_Name
),
Growth AS (
    SELECT 
        r.State_Name,
        r.Recent_Production,
        p.Past_Production,
        ROUND(((r.Recent_Production - p.Past_Production) / p.Past_Production) * 100, 2) AS Growth_Percentage
    FROM Recent r
    JOIN Past p ON r.State_Name = p.State_Name
)
SELECT *
FROM Growth
ORDER BY Growth_Percentage DESC
LIMIT 5;
'''

# Execute and fetch result into DataFrame
cursor.execute(query)
results = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]

df_growth = pd.DataFrame(results, columns=columns)

# Show the result
print(df_growth)
```

```python
import pandas as pd

# Step 1: SQL query to fetch relevant data
query = '''
SELECT
    Year,
    State_Name,
    Dist_Name,
    RICE_AREA_1000_ha,
    RICE_PRODUCTION_1000_tons,
    WHEAT_AREA_1000_ha,
    WHEAT_PRODUCTION_1000_tons,
    MAIZE_AREA_1000_ha,
    MAIZE_PRODUCTION_1000_tons
FROM agri_data
WHERE
    RICE_AREA_1000_ha IS NOT NULL AND RICE_PRODUCTION_1000_tons IS NOT NULL AND
    WHEAT_AREA_1000_ha IS NOT NULL AND WHEAT_PRODUCTION_1000_tons IS NOT NULL AND
    MAIZE_AREA_1000_ha IS NOT NULL AND MAIZE_PRODUCTION_1000_tons IS NOT NULL
'''

# Step 2: Execute query and load into DataFrame
cursor.execute(query)
results = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
df = pd.DataFrame(results, columns=columns)

# Step 3: Calculate correlation for each district
correlation_results = []

# Group by State and District
for (state, district), group in df.groupby(['State_Name', 'Dist_Name']):
    if len(group) >= 2:  # At least two years required for correlation
        rice_corr = group['RICE_AREA_1000_ha'].corr(group['RICE_PRODUCTION_1000_tons'])
        wheat_corr = group['WHEAT_AREA_1000_ha'].corr(group['WHEAT_PRODUCTION_1000_tons'])
        maize_corr = group['MAIZE_AREA_1000_ha'].corr(group['MAIZE_PRODUCTION_1000_tons'])

        correlation_results.append({
            'State_Name': state,
            'Dist_Name': district,
            'Rice_Correlation': rice_corr,
            'Wheat_Correlation': wheat_corr,
            'Maize_Correlation': maize_corr
        })

# Step 4: Create final DataFrame and sort by Rice correlation (or any)
df_correlation = pd.DataFrame(correlation_results)

# Optional: Sort by correlation strength (descending)
df_sorted = df_correlation.sort_values(
    by=['Rice_Correlation', 'Wheat_Correlation', 'Maize_Correlation'],
    ascending=False
)

# Display top correlated districts
print(df_sorted.head(10))
```

```python
##Yearly Production Growth of Cotton in Top 5 Cotton Producing States
# SQL query to calculate yearly production of cotton in top 5 states
query = '''
WITH TopCottonStates AS (
    SELECT State_Name
    FROM agri_data
    GROUP BY State_Name
    ORDER BY SUM(COTTON_PRODUCTION_1000_tons) DESC
    LIMIT 5
)
SELECT
    Year,
    State_Name,
    SUM(COTTON_PRODUCTION_1000_tons) AS Cotton_Production
FROM agri_data
WHERE State_Name IN (SELECT State_Name FROM TopCottonStates)
  AND COTTON_PRODUCTION_1000_tons IS NOT NULL
GROUP BY Year, State_Name
ORDER BY Year ASC, State_Name;
'''

# Execute and convert to DataFrame
cursor.execute(query)
results = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
df_cotton_growth = pd.DataFrame(results, columns=columns)

# Display the DataFrame
print(df_cotton_growth)
```

```python
##Districts with the Highest Rice Yield
query = '''
WITH LatestYear AS (
    SELECT MAX(Year) AS MaxYear
    FROM agri_data
)
SELECT
    Dist_Name,
    State_Name,
    Year,
    RICE_YIELD_Kg_per_ha
FROM agri_data
WHERE Year = (SELECT MaxYear FROM LatestYear)
    AND RICE_YIELD_Kg_per_ha IS NOT NULL
ORDER BY RICE_YIELD_Kg_per_ha DESC
LIMIT 10;
'''

# Execute query
cursor.execute(query)
results = cursor.fetchall()

# Get column names from cursor
columns = [desc[0] for desc in cursor.description]

# Load results into a DataFrame
df_top_rice_yield = pd.DataFrame(results, columns=columns)

# Display the top 10 districts
print(df_top_rice_yield)
```

```python
##Compare the Production of Wheat and Rice for the Top 5 States Over 10 Years
# SQL query to compare wheat and rice production for top 5 states over 10 years
query = '''
WITH TopStates AS (
    SELECT State_Name
    FROM agri_data
    GROUP BY State_Name
    ORDER BY SUM(RICE_PRODUCTION_1000_tons + WHEAT_PRODUCTION_1000_tons) DESC
    LIMIT 5
),
LatestYears AS (
    SELECT DISTINCT Year
    FROM agri_data
    ORDER BY Year DESC
    LIMIT 10
)
SELECT
    Year,
    State_Name,
    SUM(RICE_PRODUCTION_1000_tons) AS Rice_Production,
    SUM(WHEAT_PRODUCTION_1000_tons) AS Wheat_Production
FROM agri_data
WHERE State_Name IN (SELECT State_Name FROM TopStates)
  AND Year IN (SELECT Year FROM LatestYears)
GROUP BY Year, State_Name
ORDER BY Year ASC, State_Name;
'''

# Execute and convert to DataFrame
cursor.execute(query)
results = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
df_compare_production = pd.DataFrame(results, columns=columns)

# Display the DataFrame
print(df_compare_production)
```
