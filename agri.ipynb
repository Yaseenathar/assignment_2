```python
##Top 7 RICE PRODUCTION State Data(Bar_plot)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
rice_by_state = df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False)

# Select Top 7 States
top7_rice_states = rice_by_state.head(7)

# Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top7_rice_states.values, y=top7_rice_states.index, palette='viridis')
plt.title('Top 7 Rice Producing States in India')
plt.xlabel('Total Rice Production (1000 tons)')
plt.ylabel('State Name')
plt.show()
```

```python
###Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)
import matplotlib.pyplot as plt
import seaborn as sns
wheat_state =  df.groupby('State Name')['WHEAT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False)
top5_wheat_states = wheat_state.head(5)
plt.figure(figsize=(8, 8))
plt.pie(top5_wheat_states, labels=top5_wheat_states.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Percentage Contribution of Top 5 Wheat Producing States')
plt.show()
```

```python
##Oil seed production by top 5 states
import matplotlib.pyplot as plt
import seaborn as sns
oil_state = df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum().sort_values(ascending=False)
top5_oil_states = oil_state.head(5)
plt.figure(figsize = (10,6))
sns.barplot(x = top5_oil_states.index,y = top5_oil_states.values,palette='viridis')
plt.title('Oil seed production by top 5 states')
plt.xlabel('States')
plt.ylabel('Values')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sugarcane_production = df.groupby('Year')['SUGARCANE PRODUCTION (1000 tons)'].sum().reset_index()

# Sort by Year
sugarcane_production = sugarcane_production.sort_values('Year')

# Filter last 50 years if dataset has more than 50 years
if len(sugarcane_production) > 50:
    sugarcane_production = sugarcane_production.tail(50)

# Line Plot
plt.figure(figsize=(10,6))
plt.plot(sugarcane_production['Year'], sugarcane_production['SUGARCANE PRODUCTION (1000 tons)'], marker='o', color='green')
plt.title("India's Sugarcane Production Over the Last 50 Years")
plt.xlabel("Year")
plt.ylabel("Sugarcane Production (1000 tons)")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
production = df.groupby('Year')[['RICE PRODUCTION (1000 tons)', 'WHEAT PRODUCTION (1000 tons)']].sum().reset_index()

# Sort by Year
production = production.sort_values('Year')

# Filter for last 50 years if dataset is longer than 50 years
if len(production) > 50:
    production = production.tail(50)

# Line Plot for Rice vs Wheat Production
plt.figure(figsize=(10,6))
plt.plot(production['Year'], production['RICE PRODUCTION (1000 tons)'], marker='o', color='blue', label='Rice Production')
plt.plot(production['Year'], production['WHEAT PRODUCTION (1000 tons)'], marker='s', color='orange', label='Wheat Production')
plt.title("Rice Production vs Wheat Production (Last 50 Years) - India")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
wb_data = df[df['State Name'].str.upper() == 'WEST BENGAL']

# Group by Year and sum Rice Production
wb_rice = wb_data.groupby('Year')['RICE PRODUCTION (1000 tons)'].sum().reset_index()

# Sort by Year
wb_rice = wb_rice.sort_values('Year')

# Plot Rice Production trend for West Bengal
plt.figure(figsize=(10,6))
plt.plot(wb_rice['Year'], wb_rice['RICE PRODUCTION (1000 tons)'], marker='o', color='purple')
plt.title("Rice Production in West Bengal Over the Years")
plt.xlabel("Year")
plt.ylabel("Rice Production (1000 tons)")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
wb_data = df[df['State Name'].str.upper() == 'WEST BENGAL']

# Latest available year
latest_year = wb_data['Year'].max()

# Filter for latest year
wb_latest = wb_data[wb_data['Year'] == latest_year]

# Group by District and sum Rice Production (if multiple entries exist per district)
district_rice = wb_latest.groupby('Dist Name')['RICE PRODUCTION (1000 tons)'].sum().reset_index()

# Sort for better visualization
district_rice = district_rice.sort_values('RICE PRODUCTION (1000 tons)', ascending=False)

# Bar Plot
plt.figure(figsize=(12,6))
plt.bar(district_rice['Dist Name'], district_rice['RICE PRODUCTION (1000 tons)'], color='skyblue')
plt.title(f"Rice Production by Districts of West Bengal ({latest_year})")
plt.xlabel("District")
plt.ylabel("Rice Production (1000 tons)")
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
up_data = df[df['State Name'].str.upper() == 'UTTAR PRADESH']

# Group by Year and sum Wheat Production (1000 tons)
wheat_production = up_data.groupby('Year')['WHEAT PRODUCTION (1000 tons)'].sum().reset_index()

# Sort by Production descending to get Top 10 years
top_10_years = wheat_production.sort_values('WHEAT PRODUCTION (1000 tons)', ascending=False).head(10)

# Sort by Year for clean plot (optional)
top_10_years = top_10_years.sort_values('Year')

# EDA Print Summary
print("Top 10 Wheat Production Years from UP:\n")
print(top_10_years)

print("\nMaximum Production:", top_10_years['WHEAT PRODUCTION (1000 tons)'].max())
print("Minimum Production:", top_10_years['WHEAT PRODUCTION (1000 tons)'].min())
print("Average Production:", top_10_years['WHEAT PRODUCTION (1000 tons)'].mean())

# Bar Plot
plt.figure(figsize=(10,6))
plt.bar(top_10_years['Year'], top_10_years['WHEAT PRODUCTION (1000 tons)'], color='orange')
plt.title("Top 10 Wheat Production Years - Uttar Pradesh")
plt.xlabel("Year")
plt.ylabel("Wheat Production (1000 tons)")
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sorghum_data = df.groupby(['State Name', 'Year'])[
    ['KHARIF SORGHUM PRODUCTION (1000 tons)', 'RABI SORGHUM PRODUCTION (1000 tons)', 'SORGHUM PRODUCTION (1000 tons)']
].sum().reset_index()

# View summarized data
print("\nSorghum Production Summary:\n", sorghum_data.head())

# ✅ EDA 1: Latest Year - Sorghum Production by State
latest_year = sorghum_data['Year'].max()
latest_data = sorghum_data[sorghum_data['Year'] == latest_year]

# Bar Plot for Kharif and Rabi Sorghum by State
plt.figure(figsize=(14,7))
sns.barplot(x='State Name', y='KHARIF SORGHUM PRODUCTION (1000 tons)', data=latest_data, color='orange', label='Kharif')
sns.barplot(x='State Name', y='RABI SORGHUM PRODUCTION (1000 tons)', data=latest_data, color='green', label='Rabi', bottom=latest_data['KHARIF SORGHUM PRODUCTION (1000 tons)'])
plt.title(f"Sorghum Production by State - Kharif & Rabi ({latest_year})")
plt.xlabel("State")
plt.ylabel("Sorghum Production (1000 tons)")
plt.xticks(rotation=90)
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
state_groundnut = df.groupby('State Name')['GROUNDNUT PRODUCTION (1000 tons)'].sum().reset_index()

# Sort descending and pick Top 7 states
top7_states = state_groundnut.sort_values(by='GROUNDNUT PRODUCTION (1000 tons)', ascending=False).head(7)

# Show result
print("\nTop 7 States for Groundnut Production:\n", top7_states)

# ✅ Bar Plot for Top 7 States
plt.figure(figsize=(10,6))
sns.barplot(x='State Name', y='GROUNDNUT PRODUCTION (1000 tons)', data=top7_states, palette='plasma')
plt.title("Top 7 States for Groundnut Production (Total Across All Years)")
plt.xlabel("State Name")
plt.ylabel("Groundnut Production (1000 tons)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
soybean_state = df.groupby('State Name').agg({
    'SOYABEAN PRODUCTION (1000 tons)': 'sum',
    'SOYABEAN YIELD (Kg per ha)': 'mean'
}).reset_index()

# Sort by Production to get Top 5 States
top5_states = soybean_state.sort_values(by='SOYABEAN PRODUCTION (1000 tons)', ascending=False).head(5)
plt.figure(figsize=(10,6))
sns.barplot(x='State Name', y='SOYABEAN YIELD (Kg per ha)', data=top5_states, palette='magma')
plt.title("Soybean Yield Efficiency (Kg per ha) - Top 5 Producing States")
plt.xlabel("State Name")
plt.ylabel("Average Yield (Kg per ha)")
plt.grid(axis='y')
plt.show()
```

```python
oilseed_state = df.groupby('State Name')['OILSEEDS PRODUCTION (1000 tons)'].sum().reset_index()

# Sort descending to get Top 7 major states
top_states = oilseed_state.sort_values(by='OILSEEDS PRODUCTION (1000 tons)', ascending=False).head(7)

print("\nTop Oilseed Producing States:\n", top_states)

# ✅ Step 2: Bar Plot - Oilseed Production by Major States
plt.figure(figsize=(10,6))
sns.barplot(x='State Name', y='OILSEEDS PRODUCTION (1000 tons)', data=top_states, palette='rocket')
plt.title("Top 7 Oilseed Producing States (Total Production Across All Years)")
plt.xlabel("State Name")
plt.ylabel("Oilseed Production (1000 tons)")
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='RICE AREA (1000 ha)', y='RICE PRODUCTION (1000 tons)', data=df, color='green')
plt.title('Impact of Rice Cultivated Area on Production')
plt.xlabel('Rice Area (1000 ha)')
plt.ylabel('Rice Production (1000 tons)')
plt.grid()
plt.show()

# Scatter Plot: Wheat Area vs Production
plt.figure(figsize=(6,5))
sns.scatterplot(x='WHEAT AREA (1000 ha)', y='WHEAT PRODUCTION (1000 tons)', data=df, color='orange')
plt.title('Impact of Wheat Cultivated Area on Production')
plt.xlabel('Wheat Area (1000 ha)')
plt.ylabel('Wheat Production (1000 tons)')
plt.grid()
plt.show()

# Scatter Plot: Maize Area vs Production (if columns exist)
if 'MAIZE AREA (1000 ha)' in df.columns and 'MAIZE PRODUCTION (1000 tons)' in df.columns:
    plt.figure(figsize=(6,5))
    sns.scatterplot(x='MAIZE AREA (1000 ha)', y='MAIZE PRODUCTION (1000 tons)', data=df, color='blue')
    plt.title('Impact of Maize Cultivated Area on Production')
    plt.xlabel('Maize Area (1000 ha)')
    plt.ylabel('Maize Production (1000 tons)')
    plt.grid()
    plt.show()
else:
    print("⚠️ Maize area/production columns not found in dataset.")

# Correlation Coefficients
print("\nCorrelation Coefficients:")
print("Rice Area vs Production:", df['RICE AREA (1000 ha)'].corr(df['RICE PRODUCTION (1000 tons)']))
print("Wheat Area vs Production:", df['WHEAT AREA (1000 ha)'].corr(df['WHEAT PRODUCTION (1000 tons)']))

# Check for Maize if columns present
if 'MAIZE AREA (1000 ha)' in df.columns and 'MAIZE PRODUCTION (1000 tons)' in df.columns:
    print("Maize Area vs Production:", df['MAIZE AREA (1000 ha)'].corr(df['MAIZE PRODUCTION (1000 tons)']))
```
